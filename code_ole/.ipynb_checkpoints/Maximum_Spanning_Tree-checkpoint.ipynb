## Proximity

```python
edges = edge_list[["tag1","tag2","proximity"]]
edges = edges.rename(columns={"proximity": "weight"})
```

```python
# threshold: mean of proximity
edges.weight.mean()
```

```python
edges[edges["weight"] > 0.04].shape
```

```python
# reduce number of edges - threshold
edges = edges[edges["weight"] > 0.04]
```

```python
# Create graph with all edges
G=nx.from_pandas_edgelist(edges, source='tag1', target='tag2', edge_attr="weight")
```

```python
# Create maximum spanning tree
T = nx.maximum_spanning_tree(G, weight='weight')
```

```python
sorted(T.edges(data=True))
```

```python
nx.draw_spring(T)
```
